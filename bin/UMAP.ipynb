```python
import matplotlib.pyplot as plt
import params as prm
import pandas as pd
import numpy as np
import umap
import os
```

```python
df_1f = pd.read_csv(os.path.join(prm.datasets_pp_path, f"DF_{prm.dataset_1er_tour}.csv"))
df_2f = pd.read_csv(os.path.join(prm.datasets_pp_path, f"DF_{prm.dataset_2nd_tour}.csv"))
print(df_1f.shape, df_2f.shape)
```

```python
umap_1 = umap.UMAP(n_components=2).fit_transform(df_1f[df_1f.keys()[:-1]].to_numpy())
```

```python
fig, ax = plt.subplots(
    nrows=1,
    ncols=1,
    figsize=(5,5),
    dpi=150
)

ax.scatter(
    umap_1[:,0],
    umap_1[:,1],
    s=1,
    alpha=0.1,
    c=np.argmax(df_1f[df_1f.keys()[:-1]], axis=1)
)

plt.tight_layout()
plt.show()
```

```python
umap_2 = umap.UMAP(n_components=2).fit_transform(df_2f[df_2f.keys()[:-1]].to_numpy())
```

```python
fig, ax = plt.subplots(
    nrows=1,
    ncols=1,
    figsize=(5,5),
    dpi=150
)

ax.scatter(
    umap_2[:,0],
    umap_2[:,1],
    s=1,
    alpha=0.1,
    c=np.argmax(df_2f[df_2f.keys()[:-1]], axis=1)
)

plt.tight_layout()
plt.show()
```
